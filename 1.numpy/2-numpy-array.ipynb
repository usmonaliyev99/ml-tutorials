```python
# Let's create random numbers

import random

random_numbers = [random.randint(0, x) * x for x in range(100)]
random_numbers[0:10]
```

```python
# Let's convert random numbers to np array

import numpy as np

np_random_numbers = np.array(random_numbers)
np_random_numbers[0:10]
```

```python
# Let's see dimentions of array

np_random_numbers.ndim
```

```python
# Let's create a 2 dimensional array

dm_random_numbers = [ [ y * random.randint(0, x + 1) for y in range(10)] for x in range(40)]
dm_random_numbers[0:10]
```

```python
# Let's convert dm_random_numbers to numpy array

np_dm_random_numbers = np.array(dm_random_numbers)
np_dm_random_numbers.ndim
```

```python
# Let's see shape of np_dm_random_numbers

np_dm_random_numbers.shape
```

```python
# Let's see size of np_dm_random_numbers

np_dm_random_numbers.size
```

```python
# Let's use zeros function of numpy library

np_zeros = np.zeros((40, 10))
np_zeros[0:10]
```

```python
# Let's create np array with ones function

np_ones = np.ones((30,15))
np_ones[0:10]
```

```python
# Let's create array with arange function in np library

np_numbers = np.arange(2, 100, 4)
np_numbers
```

```python
# Let's use random module of np

np_rdim_numbers = np.random.rand(3, 2)
np_rdim_numbers

np_rint_numbers = np.random.randint(low=3, high=99,size=(50 , 9))
np_rint_numbers[:15]
```

```python
# Let's see full function of numpy library

np_full_numbers = np.full(shape=(32, 11), fill_value=5)
np_full_numbers[:10]
```

```python
# Let's use eye and identity function

np_eye_numbers = np.eye(N=10, M=10)
print(np_eye_numbers)

np_identity_numbers = np.identity(5)
print(np_identity_numbers)
```
