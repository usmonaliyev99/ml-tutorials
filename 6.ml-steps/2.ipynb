```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
df = pd.read_csv('https://github.com/ageron/handson-ml2/blob/master/datasets/housing/housing.csv?raw=true')

df.head()
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)

print(train_set.shape)
print(test_set.shape)
```

```python
housing_labels = train_set['median_house_value'].copy()
housing = train_set.drop(columns='median_house_value')
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')

numeric_housing = housing.drop('ocean_proximity', axis=1)

X = imputer.fit_transform(numeric_housing)

X
```

```python
X = pd.DataFrame(X, columns=numeric_housing.columns, index=numeric_housing.index)

X
```
