```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population.csv', index_col='country', usecols=['country', 'pop2021', 'area'])

df.to_csv('../datasets/world_population.csv')

df
```

```python
df.loc['Uzbekistan']
```

```python
bins = [0, 1000, 10000, 30000, 50000, 100000, 300000, 1500000]
```

```python
groups = pd.cut(df['pop2021'], bins)

groups
```

```python
groups.value_counts()
```

```python
autodf = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/automobile_data.csv', index_col='index')

autodf.to_csv('../datasets/automobile_data.csv')

autodf
```

```python
autodf.price.describe()
```

```python
auto_bins = [0, 10000, 20000, 30000, 40000, 50000]

autodf['group'] = pd.cut(autodf['price'], auto_bins)

autodf
```

```python
autodf.group.value_counts()
```

```python
auto_bins = [0, 10000, 30000, 50000]
labels = ['arzon', "o'rta", 'qimmat']

autodf['group'] = pd.cut(autodf['price'], auto_bins, labels=labels)

autodf['group'].value_counts()
```

```python
parts = pd.cut(autodf.price, 3)

parts.value_counts()
```

```python
parts = pd.qcut(autodf['horsepower'], 3)

parts.value_counts()
```
