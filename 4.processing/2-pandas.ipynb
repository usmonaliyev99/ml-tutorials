```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/MansurCompAI/Practicum_Datasets/refs/heads/master/world_cities_air_water_quality.csv')

df.head()
```

```python
df.shape
```

```python
df.columns = ['City', 'Region', 'Country', 'AirQuality', 'WaterPollution']

df.head()
```

```python
df.Region = df.Region.str.replace('"', '')

df.head()
```

```python
df.Country = df.Country.str.replace('"', '')

df.head()
```

```python
df.to_csv('../datasets/world_cities_air_water_quality.csv', index=False)
```

```python
df.Country = df.Country.str.strip()
```

```python
statistics = df.groupby(['Country', 'Region', 'City']).mean()

statistics.head(10)
```

```python
'Uzbekistan' in statistics.index
```
