```python
import pandas as pd
import numpy as np
```

```python
X = np.arange(2, 99, 4)
X
```

```python
y = X ** 2

y
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(x=X, y=y ,color='red')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr_model = LinearRegression()
```

```python
lr_model.fit(X.reshape((25, 1)), y)
```

```python
labels = lr_model.predict(X.reshape((25, 1)))

labels
```

```python
sns.scatterplot(x=X, y=y)
sns.lineplot(x=X, y=labels, color='red')

plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=False)
```

```python
X_poly = poly.fit_transform(X.reshape((25, 1)))
```

```python
lr_model.fit(X_poly, y)
```

```python
labels = lr_model.predict(X_poly)

labels
```

```python
sns.scatterplot(x=X, y=y)
sns.lineplot(x=X, y=labels, color='red')

plt.show()
```
