```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./../datasets/earth.csv', index_col='ID')

df.head()
```

```python
from datetime import datetime
```

```python
trash_indexes = df[df.Date.str.len() != 10].index
df.drop(index=trash_indexes, inplace=True)

df.Date.str.len().value_counts()
```

```python
df.Date = df.Date.apply(lambda d: datetime.strptime(d, '%m/%d/%Y'))

df.head()
```

```python
df.Time = df.Time.apply(lambda t: datetime.strptime(t, '%H:%M:%S'))
```
