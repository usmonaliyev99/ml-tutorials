```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/automobile_data_nan.csv', index_col='index')

df
```

```python
df['num-of-cylinders'].value_counts()
```

```python
numbers = {
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleventh': 11,
    'twelve': 12,
}

df['cylinders'] = df['num-of-cylinders'].map(numbers)

df.drop('num-of-cylinders', axis=1, inplace=True)

df
```

```python
def usd_to_uzs(value):
    return value * 12600

df['price'] = df['price'].map(usd_to_uzs)

df
```
