```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/automobile_data_nan.csv', index_col='index')

df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(axis=1, thresh=48)
```

```python
df.dropna(subset=['price'])
```

```python
df.fillna(0)
```

```python
df.fillna({
    'horsepower': df.horsepower.mean(),
    'average-mileage': df['average-mileage'].mean(),
})
```

```python
df.fillna(df.mean(numeric_only=True))
```

```python
df.fillna(method='ffill')
```

```python
df.fillna(method='bfill')
```

```python
audi = df[df.company == 'audi']

audi.fillna(audi.mean(numeric_only=True))
```
