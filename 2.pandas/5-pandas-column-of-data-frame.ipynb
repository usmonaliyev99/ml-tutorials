```python
import pandas as pd
```

```python
data = {
    'yil': [2021, 2020, 2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012],
    'aholi soni': [33.9, 33.1, 32.4, 32.1, 31, 30.5, 30.1, 29, 28, 27.9],
    'temp': [0.8, 0.3, 0.3, 1.1, 0.5, 0.4, 1.1, 1, 1.9, 0],
}

data
```

```python
df = pd.DataFrame(data, columns=['yil', 'temp', 'aholi soni'])

df.head()
```

```python
df = pd.DataFrame(data, columns=['yil', 'temp', 'aholi soni', 'YAIM'])

df.head()
```

```python
df = pd.DataFrame(data, index=[21, 20, 19, 18, 17, 16, 15, 14, 13, 12])

df
```

```python
df = pd.DataFrame(data, index=df['yil'].sort_values(ascending=True).values, columns=['yil', 'temp', 'aholi soni', 'yaim'])

df
```

```python
yaim = pd.Series([1800, 2341, 2341, 3213, 5643, 5332], index=[2012, 2013, 2014, 2016, 2019, 2022])

df.yaim = yaim


df.iloc[df.yaim.isnull()] = 0

df
```

```python
df['1500+'] = df.yaim > 1500

df
```
