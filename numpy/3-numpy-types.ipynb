```python
# Let's create array with float64 type

import numpy as np
import random

random_numbers = [ [ j * i for j in range(5)] for i in range(12) ]
random_numbers
```

```python
np_float64_numbers = np.array(random_numbers, dtype=np.float64)
np_float64_numbers
```

```python
np_int32_numbers = np.array(random_numbers, dtype=np.int32)
np_int32_numbers
```

```python
# Let's see type of arrays

np_float64_numbers.dtype
```

```python
# Let's see type of arrays

np_int32_numbers.dtype
```

```python
# Let's use astype function

np_float_random_numbers = np.random.random((4, 8))
np_int64_random_numbers = np_float_random_numbers.astype(dtype=np.int64)
np_int64_random_numbers.dtype
```
