```python
# Let's import numpy library

import numpy as np
```

```python
# Let's create random numbers

np_numbers = np.arange(10, 100, 10)
np_numbers
```

```python
# Let's cut some elements from array

np_sliced_numbers = np_numbers[3:5]
np_sliced_numbers
```

```python
np_sliced_numbers[:] = np_sliced_numbers * 2

np_numbers
```

```python
# Let's use copy function

np_copy_numbers = np_numbers[0:3].copy()
np_copy_numbers
```

```python
np_copy_numbers[:] = np_copy_numbers / 10

np_copy_numbers
```

```python
# Let's see np_numbers

np_numbers
```
