```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
numbers = np.arange(10)

numbers
```

```python
plt.plot((numbers))
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y, linestyle=':')
plt.show()
```

```python
x = np.arange(start=100, stop=1, step=-2)
y = x * x

plt.plot(x, y, linestyle=':')
plt.xlabel('Son')
plt.ylabel('Sonning kvadrati')
plt.show()
```

```python
x = np.arange(start=5, stop=25, step=5)

kvadrat = x ** 2
kub = x ** 3

plt.plot(x, kvadrat, linestyle=':', color='red')
plt.plot(x, kub, linestyle=':', color='blue')

plt.legend(['Kvadrati', 'Kubi'])

plt.show()
```
