```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/blob/main/uzbekistan.csv?raw=true', index_col=0)

df.columns = ['maydoni', 'aholisi']

df
```

```python
plt.figure(figsize=(10, 5))

sns.barplot(x=df.index, y=df['aholisi'])

plt.xticks(rotation=90)
plt.title("Viloyatlar aro aholi soni")
plt.ylabel('Aholi soni (million)')
plt.xlabel('Viloyatlar')
plt.grid()

plt.show()
```
