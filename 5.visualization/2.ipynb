```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/museum_visitors.csv', index_col='Date')

df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from datetime import datetime

def format(date: str):
    date: datetime = datetime.strptime(date, '%Y-%m-%d')

    return date.strftime('%d.%m.%Y')

df.index = df.index.map(format)
```

```python
plt.figure(figsize=(16, 8))

sns.lineplot(data=df)

plt.xticks(rotation=90)
plt.title("Muzeylarga tashrif soni")
plt.xlabel('Sana')
plt.ylabel('Tashriflar soni')

plt.show()
```

```python
plt.figure(figsize=(16, 5))

sns.lineplot(data=df[['America Tropical Interpretive Center', 'Chinese American Museum']])

plt.xticks(rotation=90)

plt.show()
```
