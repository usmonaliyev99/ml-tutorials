# Scatter plot

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./../datasets/merc.csv')

df.head()
```

```python
df = df[df.year > 2000]
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(df, x='year', y='price')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.regplot(df, x='year', y='price')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.regplot(df, x='engineSize', y='price')

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(df, x='year', y='price', hue='fuelType')

plt.show()
```
