```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./../datasets/merc.csv')

df.head()
```

```python
sns.displot(data=df, x='price', kind='kde', aspect=2, height=5)

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.kdeplot(data=df, x='price', hue='fuelType', fill=True)

plt.show()
```
