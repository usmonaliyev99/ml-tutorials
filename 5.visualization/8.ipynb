```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./../datasets/merc.csv')

df.head()
```

```python
df = df[(df.year >= 2010) & (df.transmission != 'Other') & (df.fuelType != 'Other')]

df.shape
```

```python
sns.histplot(data=df, x='price')

plt.show()
```

```python
sns.scatterplot(data=df, x='year', y='price')

plt.show()
```

```python
fig, ax = plt.subplots(2, 1, figsize=(10, 6))

sns.histplot(data=df, x='price', ax=ax[0])
sns.scatterplot(data=df, x='year', y='price', ax=ax[1])

plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 10))

sns.lineplot(data=df, x='year', y='price', ax=ax[0, 0])
sns.barplot(data=df, x='year', y='price', ax=ax[0, 1])
sns.scatterplot(data=df, x='year', y='price', ax=ax[1, 0])
sns.histplot(data=df, x='price', ax=ax[1, 1])

plt.show()
```
