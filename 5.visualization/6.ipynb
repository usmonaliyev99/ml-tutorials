```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./../datasets/merc.csv')

df.head()
```

```python
df.info()
```

```python
df = df[(df.fuelType != 'Other') & (df.transmission != 'Other') & (df.year >= 2010)]

df.shape
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(data=df, x='price')

plt.show()
```

```python
sns.displot(data=df, x='price', aspect=2)

plt.show()
```

```python
sns.displot(data=df, x='price', hue='fuelType', aspect=2, multiple='stack')

plt.show()
```

```python
sns.displot(df, x='price', col='fuelType', aspect=2)

plt.show()
```
